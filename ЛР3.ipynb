```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y1, linestyle='-', marker='o', markevery=20, label='sin(x)')
ax.plot(x, y2, linestyle='--', marker='s', markevery=20, label='cos(x)')

ax.set_title("Дві лінії")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.5)
ax.legend()
plt.show()
```

```python
np.random.seed(1)
x = np.random.randn(200)
y = 1.2 * x + np.random.randn(200) / 2

c = np.hypot(x, y)

fig, ax = plt.subplots(figsize=(6, 3))
sc = ax.scatter(x, y, c=c, s=60, edgecolor='k')
ax.set_title("Scatter: колір за величиною")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.colorbar(sc, ax=ax, label="шкала кольору")
ax.grid(True, alpha=0.3)

plt.show()
```

```python
cats = ["A", "B", "C", "D", "E", "F"]
vals = [22, 8, 39, 2, 11, 19]

fig, ax = plt.subplots(figsize=(6, 3))
bars = ax.bar(cats, vals)

ax.set_title("Bar chart із підписами")
ax.set_xlabel("Категорія")
ax.set_ylabel("Значення")
ax.grid(axis='y', alpha=0.3)

for b in bars:
    ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.3,
            f"{b.get_height():.0f}", ha="center", va="bottom")

plt.show()
```

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import matplotlib.gridspec as gridspec

# Налаштування даних
np.random.seed(6)
t_idx = np.arange(20)
data1 = np.random.randn(20)
data2 = np.random.randn(20) + 0.5
data3 = np.random.randn(20) - 0.5

# Дані для контуру
Xc, Yc = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Zc = (1 - Xc/2 + Xc**5 + Yc**3) * np.exp(-Xc**2 - Yc**2)

# Дані для 3D
X3, Y3 = np.meshgrid(np.arange(-5, 5, 0.25), np.arange(-5, 5, 0.25))
R3 = np.sqrt(X3**2 + Y3**2)
Z3 = np.cos(R3)

# Створення фігури та сітки
fig = plt.figure(figsize=(16, 10), constrained_layout=True)
gs = gridspec.GridSpec(2, 2, figure=fig)

# 1. Лінійний графік (Зверху зліва)
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(t_idx, data1, label="data1", color='tab:blue')
ax1.plot(t_idx, data2, label="data2", color='tab:orange')
ax1.plot(t_idx, data3, 'd', label="data3", color='tab:green')
ax1.set_title("Легенда та маркери", fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Контурний графік (Зверху справа)
ax2 = fig.add_subplot(gs[0, 1])
levels = np.linspace(Zc.min(), Zc.max(), 10)
contour = ax2.contourf(Xc, Yc, Zc, levels=levels, cmap='coolwarm')
fig.colorbar(contour, ax=ax2)
ax2.set_title("Контурне поле (Coolwarm)", fontsize=14)

# 3. 3D Поверхня (Знизу - на всю ширину)
ax3 = fig.add_subplot(gs[1, :], projection='3d')
surf = ax3.plot_surface(X3, Y3, Z3, cmap=cm.viridis, linewidth=0,
                        antialiased=False, alpha=0.8)
ax3.set_title("3D Поверхня (Viridis)", fontsize=14)
ax3.set(xticklabels=[], yticklabels=[], zticklabels=[])
fig.colorbar(surf, ax=ax3, shrink=0.5, aspect=10)

# Загальний заголовок
fig.suptitle('Комплексний візуальний дашборд', fontsize=22, fontweight='bold')

plt.show()

```python

np.random.seed(6)
t_idx = np.arange(20)
d1, d2, d3 = np.random.randn(20), np.random.randn(20) + 0.5, np.random.randn(20) - 0.5

Xc, Yc = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Zc = (1 - Xc/2 + Xc**5 + Yc**3) * np.exp(-Xc**2 - Yc**2)

ta = np.arange(0.0, 5.0, 0.01)
sa = np.cos(2 * np.pi * ta)

X3, Y3 = np.meshgrid(np.arange(-5, 5, 0.25), np.arange(-5, 5, 0.25))
R3 = np.sqrt(X3**2 + Y3**2)
Z3 = np.cos(R3)

fig = plt.figure(figsize=(16, 12), constrained_layout=True)
gs = gridspec.GridSpec(2, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(t_idx, d1, label="data1", alpha=0.8)
ax1.plot(t_idx, d2, label="data2", alpha=0.8)
ax1.plot(t_idx, d3, 'd', label="data3", markersize=4)
ax1.set_title("1. Ряди даних", fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2 = fig.add_subplot(gs[0, 1])
levels = np.linspace(Zc.min(), Zc.max(), 10)
contour = ax2.contourf(Xc, Yc, Zc, levels=levels, cmap='coolwarm')
fig.colorbar(contour, ax=ax2)
ax2.set_title("2. Контурний аналіз", fontsize=14, fontweight='bold')

ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(ta, sa, linewidth=2, color='tab:red')
ax3.annotate("local max", xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor="black", shrink=0.05))
ax3.set_title("3. Анотація тригонометрії", fontsize=14, fontweight='bold')
ax3.set_ylim(-2, 2)
ax3.grid(True, alpha=0.3)

ax4 = fig.add_subplot(gs[1, 1], projection='3d')
surf = ax4.plot_surface(X3, Y3, Z3, cmap=cm.viridis, antialiased=False, alpha=0.8)
ax4.set_title("4. Математична поверхня", fontsize=14, fontweight='bold')
fig.colorbar(surf, ax=ax4, shrink=0.5, aspect=10)

fig.suptitle('Фінальний аналітичний дашборд Matplotlib', fontsize=24, fontweight='bold', y=1.05)

plt.show()
```

```python

from matplotlib.animation import FuncAnimation
from IPython.display import HTML

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_title("Анімація Точка на синусоїді", fontsize=14)

ax.plot(x, y, color='blue', alpha=0.2, lw=2)

dot, = ax.plot([], [], 'ro', markersize=10, zorder=5)
coord_text = ax.text(0, 0, '', fontsize=11, fontweight='bold',
                     bbox=dict(facecolor='white', alpha=0.7, edgecolor='red'))

def init():
    dot.set_data([], [])
    coord_text.set_text('')
    return dot, coord_text

def update(frame):
    curr_x = x[frame]
    curr_y = y[frame]

    dot.set_data([curr_x], [curr_y])

    coord_text.set_text(f'x={curr_x:.2f}\ny={curr_y:.2f}')
    coord_text.set_position((curr_x + 0.2, curr_y + 0.2))

    return dot, coord_text
ani = FuncAnimation(fig, update, frames=len(x), init_func=init,
                    blit=True, interval=50)

plt.close()

HTML(ani.to_html5_video())
```
